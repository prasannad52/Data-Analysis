```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,4 * np.pi)
sin_y = np.sin(x)
cos_y = np.cos(x)
plt.plot(x,sin_y,label = 'sin')
plt.title('sine and cos wave')
plt.plot(x,cos_y,label = 'cosine')
plt.legend(loc = 'upper right'  )
plt.show()
```

```python
x = np.linspace(0,4 * np.pi,1000)
sin_y = np.sin(x)
cos_y = np.cos(x)
plt.plot(x,sin_y,'*')
plt.title('sine and cos wave')
plt.plot(x,cos_y,'--')
plt.legend(loc = 'upper right'  )
plt.show()
```

```python
plt.figure(figsize = (10,5))
x = np.linspace(0,4 * np.pi,100)
plt.style.use('Solarize_Light2')
sin_y = np.sin(x)
plt.plot(x,sin_y)
plt.title('sine and cos wave')
plt.show()
```

```python
# xlim and ylim function
plt.figure(figsize = (10,5))
x = np.linspace(0,4 * np.pi,100)
plt.style.use('Solarize_Light2')
sin_y = np.sin(x)
plt.xlim([0,10])
plt.ylim([-2,2])
plt.plot(x,sin_y)
plt.title('sine and cos wave')
plt.show()
```

```python
x = np.linspace(0, 3*np.pi, 100)

fig = plt.axes()

fig.plot(x, np.sin(x))

fig.set(xlim = (0,10), ylim = (-2,2), xlabel = 'x', ylabel = 'sin(x)',title = 'sine wave')

fig.plot()
```

```python
plt.figure(figsize = (20,15))

x = np.random.randn(10000)

y = np.random.randn(10000)

plt.scatter(x,y,color='g',alpha = 1,s=10)

plt.title('Scatter plot')

plt.show()
```
